## More on confounding : Do doctors commit more sexual harressment?

![sexcrime](Images/crimestat.png)
*Figure: National crime statistics*

According to the national crime statistic, doctors were ranked the second most job which committed sexual harressment.
One might think that being a doctor is correlated to committing sexual crimes.
However, this is a typical example of confounding.
One can describe the current problem in a form of _Directed Acyclic Graph_ (DAGs) as follows.

![dagsex](Images/dagsex.png)
*Figure: DAG of being a doctor and committing a crime (Generated by DAGitty v2.3)*

Using the data from the national crime statistic, the below code shows that doctors aren't more dangerous than the general population in terms of sex crimes.

```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
import colour as colour

# Parameter Data: 2017 기준 KOSIS 여성범죄자 통계, 보건통계연보, 검찰청 범죄통계
data_pop = pd.DataFrame([[51422507,25768055,25654452],[108000,79000,29000]],
                        index=['general', 'doctor'],
                         columns=['total', 'men', 'women'])
data_crime = pd.DataFrame([[32837,31905,829],[138,138,0]],
                        index=['general', 'doctor'],
                         columns=['total', 'men', 'women'])
data_cr = data_crime / data_pop * 100000


# Draw figures : Crime rate
x = np.arange(len(data_pop.index)) * 2
labels = ['General Population', 'Doctors']
width = 0.2
fig, ax = plt.subplots()
rects1 = ax.bar(x-width, data_cr['total'], width, label='Total', color='grey')
rects2 = ax.bar(x, data_cr['men'], width, label='Men', color='skyblue')
rects3 = ax.bar(x+width, data_cr['women'], width, label='Women', color='pink')

ax.set_ylabel('Sex crimes per 100,000')
ax.set_ylim([0,200])
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_xlim([-1,3])
ax.legend()

# label doctor/general ratio on top
for dx, ycor, num in zip([-width,0,width], data_cr.loc['doctor',:], data_cr.loc['doctor',:]/data_cr.loc['general',:]):
    ax.text(x=x[1]+dx, y=ycor+5, ha='center', s='%.3g' % num, fontsize=15)

plt.show()


```

The labels on top of the bars are the crime rate of doctors divided by the crime rate of the general population.
After stratifying sex, the incidence of sex crimes droped from 2 to 1.41, which means that approximately 60% of increased sex crime incidences by doctors are explained by the sex ratio among doctors.
### From this we see that by increasing the number of female doctors, crime rate of doctors will decrease without any other intervention.
